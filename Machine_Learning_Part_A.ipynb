```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = [2,5,7,3,1,4,6,8,9,13,4,8,10,12]
y= [12,34,23,10,3,45,67,90,24,48,30,46,54,60]

slope, intercept, r, p, std_err = stats.linregress(x,y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc,x))

plt.scatter(x,y)
plt.plot(x, mymodel)
plt.show()

print(r) # A coefficient of 0.6 indicates that there is a relationship between x and y but not that strong though it supports a linear prediction

print() # Used to add a blank line in between the output of the code

speed = myfunc(11)
print(speed)

print()

print("The value of p is:", p)
print()
print("The value of std_err is:", std_err)
print()
print("The intercept value is:", intercept)
print()
print('The value of the slope is:', slope)
print()
```

```python
DATA = {
    "Age" : [36, 42, 23, 52, 43, 44, 66, 35, 52, 35, 24, 18,45],
    "Experience" : [10, 12, 4, 4, 21, 14, 3, 14, 13, 5, 3, 3, 9],
    'Rank' : [9, 4, 6, 4, 8, 5, 7, 9, 7, 9, 5, 7, 9],
    'Nationality' : ['UK', 'USA', 'N', 'USA', 'USA', 'UK', 'N', 'UK', 'N', 'N', 'USA', 'UK', 'UK'],
    'Go' : ['NO', 'NO', 'NO', 'NO', 'YES', 'NO', 'YES', 'YES', 'YES', 'YES', 'NO', 'YES', 'YES']
}

df = pd.DataFrame(DATA)
print(df)
```

Since we want to use the above data set in creating a Decision Tree, we'll need to make all the entries to be in numerical form.

```python
Natnal_change = {'UK' : 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(Natnal_change)

Go_change = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(Go_change)

print(df)
```

Now that all the entries are in the numerical form, we'll split the features and the target data sets

```python
features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

print('Our features are:\n', X)

print()

print('Our target is:\n', y)
```

Let's now create the Decision Tree

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

dcsn_tree = DecisionTreeClassifier()
dcsn_tree = dcsn_tree.fit(X,y)

tree.plot_tree(dcsn_tree, feature_names= features)
```

NB : The Decision Tree does not give us a **certain** answer. This is true since whenever we run the code, it gives us different answers even if we're still using the same data. It uses the probability concept for any outcome.

GINI = 1 - [(a/n)^2 - (b/n)^2] where n is the number of samples, a is eg the +ve values(YES) and b is eg the -ve values(NO)

Through the 'Rank' feature, all the entries in the 'Rank' that have the value equal or less than 6.5 will be shifted to the LHS and the converse to the RHS

```python
print(dcsn_tree.predict([[40, 10, 7, 1]])) # Using the Decision Tree Model to make predictions
print()
```

The above output is the prediction of whether one should 'Go' to a concert where the comedian is 40yrs old, has 10yrs of experience, ranked 7 and one that comes from USA. The prediction is [1] which means YES according to the way we had tranformed the entries of our Target variable 'Go'

```python
from sklearn import metrics

actual = np.random.binomial(1, 0.9, size = 1000) # The Binomial distribution
predicted = np.random.binomial(1, 0.9, size = 1000)

confsn_matrix = metrics.confusion_matrix(actual, predicted)

show_the_matrix = metrics.ConfusionMatrixDisplay(confusion_matrix = confsn_matrix , display_labels= [0, 1])

show_the_matrix.plot()
plt.show()
```

The output keeps changing once the code is re-run. What we need to note is that the yellow part cell is called the **TP**, the cell above it is called **FP**, the cell horizontally adjacent to it(the yellow part) is called **FN** while the other cell is the **TN**.

```python
Accuracy = metrics.accuracy_score(actual, predicted)
print('The accuracy of the matrix is : ' , Accuracy )
print()

Precision = metrics.precision_score(actual, predicted)
print('The precision of the matrix is : ' , Precision)
print()

Recall = metrics.recall_score(actual, predicted) # Recall is also termed as 'sensitivity'
print('The recall of the matix is : ' , Recall)
print()

Specificity = metrics.recall_score(actual, predicted, pos_label = 0) # NB: Abt the pos_label. This metric is somehow calculated as recall
print('The specificity of the matrix is : ' , Specificity)
print()

F1_score = metrics.f1_score(actual, predicted)
print('The F1_score of the matrix is : ' , F1_score)
print()
```

The above oututs represents some of the metrics of evaluating models but to be specific, the confusion matrix at the moment. The formulaes for calculating the metrics are as follows:

a) Accuracy = (TP + TN) / All the values

b) Precision = TP / (TP + FP) ie focuses on the outward positivity

c) Recall / sensitivity = TP / (TP + FN) ie focuses on the inward positivity

d) Specificity = TN / (TN + FP) ie focuses on the inward negativity

e) F1 Score = (2 * (Recall + Precision) ) / (Recall + Precision)
